```python
import seaborn as sns
import pandas
import numpy as np
```

```python
sns.get_dataset_names()
```

```python
penguins = sns.load_dataset('penguins')
penguins.head()
```

```python
penguins["species"].value_counts()
```

```python
penguins["island"].value_counts()
```

```python
# scatter plot:

sns.scatterplot(data=penguins, x="flipper_length_mm", y="body_mass_g", hue = "island")
```

```python
# The ways of styling themes are as follows:
# white
# dark
# whitegrid
# darkgrid
# ticks


# set_context() allows us to override default parameters. This affects things like the size of the labels, lines, and other elements of the plot, but not the overall style -> set_style(). The context are:
# poster
# Paper - Ideal for printing in academic papers.
# Notebook - Good for working inside Jupyter notebooks (default setting).
# Talk   Good for presentation slides where readability matters.
```

```python
sns.set_style("ticks")
sns.set_context("notebook")
sns.scatterplot(data=penguins, x="flipper_length_mm", y="body_mass_g", hue = "island", palette = "Dark2")
sns.despine()
```

```python
sns.scatterplot(data = penguins, x ="species", y = "body_mass_g", hue = "island", style = "sex", alpha = 0.5)
```

```python
sns.stripplot(data = penguins, x = "species", y = "body_mass_g", hue = "island", dodge = True, jitter = True)
```

```python
# Swarm plot
sns.set_context("talk")
sns.swarmplot(data = penguins, x = "species", y = "body_mass_g", hue = "island")
sns.despine()
```

```python
# histogram:
sns.set_context("notebook")
sns.histplot(data = penguins, x = "body_mass_g", hue="sex", multiple="stack")
```

```python
# regression plot:
sns.regplot(data = penguins, x = "body_mass_g", y = "flipper_length_mm", color = "green")
```

```python
# line plot:
sns.set_style("whitegrid")
sns.lineplot(data = penguins, x = "body_mass_g", y = "flipper_length_mm", hue="island", style = "sex")
```

```python
# jointplots

sns.jointplot(data = penguins, x = "body_mass_g", y = "flipper_length_mm")
```

```python
sns.jointplot(data = penguins, x = "body_mass_g", y = "flipper_length_mm", kind = "reg")
```

```python
# bar plot:

sns.barplot(data = penguins, x = "species", y = "body_mass_g", hue = "sex", palette = ["pink", "skyblue"], estimator = np.var)
```

```python
# countplot:

sns.countplot(data = penguins, x = "species", hue = "island", palette = "Set2")
```

```python
# .box plot:

sns.boxplot(data = penguins, x = "species", y = "body_mass_g", hue = "sex")
```

```python
# violin plot:

sns.violinplot(data = penguins, x = "species", y = "body_mass_g", hue = "sex", split= True, inner = "box")
```

```python
# Width of the violin: Density of the data at different values
# Middle line: Median
# Thick bar: Interquartile range (if inner='box')
```

```python
sns.violinplot(data = penguins, x = "species", y = "body_mass_g", palette = "pastel")
sns.swarmplot(data = penguins, x="species", y="body_mass_g", hue = "island", size=3, palette= "dark")
```

```python
# kde plot:
sns.kdeplot(data = penguins, x = "body_mass_g", hue = "species", fill = True)
```

```python
# Heatmap:

penguins.head()
```

```python
columns = ["bill_length_mm",	"bill_depth_mm",	"flipper_length_mm",	"body_mass_g"]

penguins[columns].corr()
```

```python
sns.heatmap(data = penguins[columns].corr(), annot=True, cmap="Blues", vmin = -0.2, linewidth = "2", linecolor = "black")
```

```python
# Rug plot:

sns.rugplot(data = penguins,  y = "body_mass_g", hue = "species", palette = "Set2", height = 0.2)
```

```python
# pair plot:

sns.pairplot(data = penguins, hue = "sex", palette = "Set2", diag_kind = "hist")
```

```python
# pair grid -> subplot grid for plotting pairwise relationships in a dataset.

graph = sns.PairGrid(data = penguins, hue = "sex", palette = "Set2")
graph.map_upper(sns.scatterplot) #upper triangle
graph.map_lower(sns.kdeplot) #lower triangle
graph.map_diag(sns.histplot) #diagnol
graph.add_legend()
```
